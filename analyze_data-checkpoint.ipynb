# Code smell Survival Analysis

```python
import json
import pandas as pd
from lifelines import CoxPHFitter
from lifelines import KaplanMeierFitter
from lifelines import NelsonAalenFitter
import matplotlib.pyplot as plt
```

## Data Frame 
Create the data frame using SZZ algoritm file and SonarQube code smell file

```python
def readData():
    data = {}

    szz_f = open('./szz/bower_SZZ_data.json')
    code_smell_f = open('./SonarQube_smells/bower_project_code_smells.json')


    szz_data = json.load(szz_f)
    code_smell_data = json.load(code_smell_f)

    for item in szz_data:
        path = item['path']
        if path not in data:
            data[path] = {"time":item['time'], 'smells':[]}
        else:
            data[path]['time'] = item['time']    

    for item in code_smell_data:
        path = item['path']
        if path not in data:
            data[item['path']] = {"time":None, 'smells':item['smells']}
        else:
            data[path]['smells'] = item["smells"]   
    return data
```

```python
def formatData():
    data = readData()
    print(data)
    time = []
    smells = []
    event =[]

    for path, info in data.items():
        time.append(info['time']) if info['time'] is not None else time.append(0)
        smells.append(1) if len(info['smells']) >0  else smells.append(0)

            event.append(0)  
        else:
            event.append(1)
        
        
    cox_data = {
        'time' : time,
        'smells':smells,
        'event':event
    }
    print("Length of each array : " )
    print("Time: " + str(len(time)) +" Smells: " + str(len(smells)) + " Event: " + str(len(event)) )
    print(cox_data)
    return cox_data
```

```python
data = formatData()
df = pd.DataFrame(data)

# cox = CoxPHFitter()
# cox.fit(df, duration_col='time', event_col='event')
# cox.print_summary()   

df_smell = df[df['smells'] == 1]
# # ###  
kmf = KaplanMeierFitter()
kmf.fit(durations=df_smell['time'], event_observed=df_smell['event'])
kmf.survival_function_.plot()
plt.title('Survival function')
# Get the current x and y axis limits
x_limits = plt.xlim()
y_limits = plt.ylim()

# Draw a horizontal line in the middle of the y-axis range
middle_y = (y_limits[0] + y_limits[1]) / 2
plt.axhline(middle_y, color='gray', linestyle='--')

# Draw a vertical line in the middle of the x-axis range
middle_x = (x_limits[0] + x_limits[1]) / 2
plt.axvline(middle_x, color='gray', linestyle='--')

# from lifelines.utils import median_survival_times
# median_ = kmf.median_survival_time_

# median_confidence_interval_ = median_survival_times(kmf.confidence_interval_)
# print(median_)
# print(median_confidence_interval_)
```

```python
# Extract survival function
survival_function = kmf.survival_function_
times = survival_function.index
survival_probs = survival_function['KM_estimate']

# Convert to numpy arrays
times = times.values
survival_probs = survival_probs.values

# Calculate the differences
delta_x = times[1:] - times[:-1]  # Change in time intervals
delta_y = survival_probs[1:] - survival_probs[:-1]  # Change in survival probabilities

# Calculate slopes
slopes = delta_y / delta_x

# Plot survival function
kmf.survival_function_.plot()
plt.title('Survival Function')
plt.xlabel('Time')
plt.ylabel('Survival Probability')
plt.show()
```
