### The proportion of the 20 violated SonarQube code smell rules

```python
data = json.load(open('./top_20_code_smells.json','r'))
js_ts_data = {key: value for key, value in data.items() if key.startswith('javascript:') or key.startswith('typescript:')}
sorted_data = sorted(js_ts_data.items(), key=lambda x: x[1], reverse=True)

top_20 = sorted_data[:20]
print(top_20)
```

# Code smell Survival Analysis

## Data Frame 
Create the data frame using SZZ algoritm file and SonarQube code smell file

```python
def readData():
    data = {}

    ## Need to change the file names
    szz_f = open('./szz/node_SZZ_data.json')
    code_smell_f = open('./SonarQube_smells/nodejs_project_code_smells.json')


    szz_data = json.load(szz_f)
    code_smell_data = json.load(code_smell_f)

    for item in szz_data:
        path = item['path']
        if path not in data:
            data[path] = {"time":item['time'], 'smells':[]}
        else:
            data[path]['time'] = item['time']    

    for item in code_smell_data:
        path = item['path']
        if path not in data:
            data[item['path']] = {"time":None, 'smells':item['smells']}
        else:
            data[path]['smells'] = item["smells"]  
    return data
```

```python
def formatData():
    data = readData()
    
    time = []
    smells = []
    event =[]

    for path, info in data.items():
        time.append(info['time']) if info['time'] is not None else time.append(0)
        smells.append(1) if len(info['smells']) >0  else smells.append(0)
        
        if info['time'] == None:
            event.append(0)  
        else:
            event.append(1) 
            
    cox_data = {
        'time' : time,
        'smells':smells,
        'event':event
    }
    print("Length of each array : " )
    print("Time: " + str(len(time)) +" Smells: " + str(len(smells)) + " Event: " + str(len(event)) )
    return cox_data
```

```python
data = formatData()
df = pd.DataFrame(data)
```

## Kaplan-Meier Estimator 
- Estimates the survival probability over time based on observed lifeitme data
- Survival posibility of each code smells
- Draw a survival curve of the overall data

```python
df_smell = df[df['smells'] == 1]
kmf = KaplanMeierFitter()
kmf.fit(durations = df_smell['time'], event_observed=df_smell['event'])
kmf.survival_function_.plot()

x_limits = plt.xlim()
y_limits = plt.ylim()

middle_y = (y_limits[0] + y_limits[1]) / 2
plt.axhline(middle_y, color='gray', linestyle='--')

middle_x = (x_limits[0] + x_limits[1]) / 2
plt.axvline(middle_x, color='gray', linestyle='--')

plt.title('Survival function')
```
